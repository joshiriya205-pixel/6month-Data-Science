```python
class School:
    def __init__(self):   # constructor
        self.name="riya"   #public variable
        self._branch="CSE"   #protected variable
        self.__rollnumber=28 #private variable  

    def Show(self):                       # method to access private variable
        print("roll number:",self.__rollnumber)



class Teacher(School):
    def __init__(self): 
        super().__init__()   # here Super() is used to call parent class constructor
        
    def display(self):
        print(self.name)
        print(self._branch)
        self.Show()  #  private variable ko print krvane k liye show() method ko call krvaya gya hai


    

student = Teacher()

student.display()

# print(student.branch)



```

```python
class School:
    def __init__(self,name,branch,rollnumber):
        self.name=name    #public variable
        self._branch=branch   #protectedd variable
        self.__rollnumber=rollnumber  #public variable

    def  show(self):
        print("Roll number:",self.__rollnumber)

class Teacher(School):
    def __init__(self,name,branch,rollnumber):
        super().__init__(name,branch,rollnumber)

    def display(self):
        print(self.name)
        print(self._branch)

        self.show()

student=Teacher(name="riya",
                branch="CSE",
                rollnumber=28)
student.display()
        
```
