#  ♦️ **Looping**
* Looping in python means repeating a block of code multiple time.

### Types of Looping 👉
1. For Loop
2. While Loop 

   
### ⏺️ **For Loop**
* A for loop in Python is used to repeat a block of code multiple times by iterating over a sequence like a list, tuple,  string, or range.

```python
name="riya"
for x in name:
    print(x)
```

```python
for x in range(1,11):
    print(x)
```

```python
for x in range(2,22,2):     #here (start,end,gap)
    print(x)
```

```python
student = {
    "name": "Riya",
    "age": 21,
    "course": "Python",
    "city": "Jaipur",
    "college": "XYZ College",
    "marks": 85,
    "grade": "A",
    "phone": 9876543210,
    "email": "ritik@gmail.com",
    "gender": "Male",
    "state": "Rajasthan",
    "country": "India",
    "language": "Hindi",
    "hobby": "Coding",
    "status": "Active"
}

print(student)
```

```python

ky=input("enter your key")
val=input("enter your value")

for key in student.keys():
    if key == ky:
        if student[key]==val:
            print("value",val)
        else:
            student[ky]=val
            print(student)
        
    else:
        print("not found")
```

```python
for x in range(2,20):
    if x==10:
        break
    print(x)
```

```python
for x in range(2,20):
    if x==10:
        continue     # it is used for skip
    print(x)
```
