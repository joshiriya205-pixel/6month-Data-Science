```python
# Write a Python program to check a student's result and scholarship eligibility using nested if statements.
student_marks = int(input("Enter your marks: "))
student_attend = int(input("Enter your attendance: "))

if student_attend >= 75:
    
    if student_marks >= 35:
        print("You are Pass")
        
        if student_marks >= 50:
            print("You can get scholarship")
        else:
            print("You are pass but not eligible for scholarship")
    
    else:
        print("You are Fail")

else:
    print("Attendance is less than 75%")
```

```python
# Write a Python program to calculate discount on a product.
items = int(input("Enter number of items: "))
price = int(input("Enter total price: "))

discount = (price * 10) / 100
total = price - discount

if items >= 5:
    print("You will get discount")
    print("Final price:", total)
else:
    print("No discount")
    print("Total price:", price)
```
