```python
#  find maximum number using DSA
def find_max(lst):
    maximum=lst[0]
    for x in lst:
        if x>maximum:
            maximum=x
    print(maximum)
find_max([1,2,3,4,5,6])
```

```python
# search element using DSA
def linear_search(arr,key):   #here arr is list and key jo value search krni hai

    for i in range(len(arr)): #len(arr):-find total length of list 
        if arr[i]==key:
            return i
    return -1
print(linear_search([10,30,50,60,80] ,80))   

```

```python
def fibonacci(n):   # here fibonacci mean check previous 2 numbers sum
    a=0
    b=1

    for i in range(n):
        print(a)

        c=a+b
        a=b
        b=c

fibonacci(5)    # 0, 0+1=1,1+1=2,1+2=3
```

```python
# find factorial using dsa
def factorial(n):

    if n==1:
        return 1
    
    return n*factorial(n-1)

print(factorial(5))
```

```python
def max_number(lst):
    maximum=lst[0]
    for i in lst:
        if i>maximum:
            maximum=i
    print(maximum)
max_number([100,2,3,4,15])
```

```python
def find_sum(lst):
    total=0
    for x in lst:
        total+=x
    return total
find_sum([1,2,3,4,5])
```

```python
def even_odd(lst):
    
    for x in lst:
        if x%2==0:
            print("even", x)
        else:
            print("odd",x)
    
even_odd([1,2,3,4,5,6,7,8,11])
```

```python
def prime_num(num):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                print("it is not a prime number",num)
                break
        else:
            print("it is prime number",num)
numbers = [10, 11, 12, 13, 14, 15, 17]

# Checking prime numbers from list
for n in numbers:
    if prime_num(n):
        print(n)
```
