```python
lst=[1,2,3,4,5,6]
def even_odd(lst):
    print("This is my list :-",lst)
    for x in lst:
        if x%2==0:
            print("even:-",x)
        else:
            print("odd:-",x)

even_odd(lst=lst)
```

```python
def fact(a):
    i=1
    while a >=1:
        i=i*a
        a-=1
    print(i)
fact(5)
```

```python

def cal_list(lst):
    total=0
    for x in lst:
        total+=x
    return total        #
cal_list(lst=[1,2,3,4,5])
    
```
