# Polymorphism


```python
# function polymorphism
print(len("hello"))
print(len([1,2,3,4]))
```

```python
#method polymorphism
class College: 
    def show(self):
        print("hello hii kese ho")
    def show(self):
        print("haan me thik hu")

College().show()

```

```python
#method polymorphism
class College: 
    def show(self):
        print("hello hii kese ho")

class Student:
    def show(self):
        print("haan me thik hu")

College().show()
Student().show()
```

```python
#operator polymorphism
print(4+1)
print("hello "+" hii")
print([1,2,3,4,4,5]+[1,3,4])
```

```python
#method overriding
class Company:
    def details(self):
        print("sb log kam kr rahe hai")
    def details(self):
        print("developer is working")

Company().details()
```
