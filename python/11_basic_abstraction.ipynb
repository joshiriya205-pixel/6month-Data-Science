```python
from abc import ABC, abstractmethod
class School(ABC):
    @abstractmethod
    def display(self):
        print("hello")
    def show(self):
        print("hii")

class Teacher(School):
    def display(self):
        print("hello")

    def show(self):
        print("hii")

# obj=School()
tc=Teacher()
tc.display()
tc.show()



```

```python
from abc import ABC, abstractmethod
class School(ABC):
    @abstractmethod
    def show(self):
        pass
    def display(self):
        print("I am not hidden")

class Student(School):
    def show(self):
        print("Abstract method complete")

# obj=School()
hello=Student()
hello.show()
hello.display()




```
