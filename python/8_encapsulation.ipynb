# Encapsulation



```python
class College:
    __college_library="yes" ## protected variable (double underscore)
    _account_information="12715365" ##private variable (single  underscore)
    college_name="CIT" ## public variable

cr =College()
# cr.__college_library
cr.account_information
cr.college_name

```

```python
# we can access private variable using this way
class College:
    __collegeaccount=45451
cr=College()
```

```python
cr.__collegeaccount
```

```python
cr._College__collegeaccount
```

```python
class Company:
    _company_account_details=5355625762153425
    def private_details(self):
        return self._company_account_details
com=Company()
com._Company__company_account_details
```
