# ♦️**Operators in Python**
* An operator is a symbol that performs a specific operation on one or more values or variables.
   
#### python operators name 👇   
1. Arithmetic Operators
2. Assignment Operators
3. Comparison Operators
4. Logical Operators
5. Bitwise Operators
6. Membership Operators
7. Identity Operators

### ⏺️ Arithmetic Operators
* Used to do mathematical calculations like addition, subtraction, multiplication, etc.
#### Arithmetic Operators table
| Operator | Name                | Example   | Output |
| -------- | ------------------- | --------- | ------ |
| `+`      | Addition            | `10 + 5`  | `15`   |
| `-`      | Subtraction         | `10 - 5`  | `5`    |
| `*`      | Multiplication      | `10 * 5`  | `50`   |
| `/`      | Division            | `10 / 5`  | `2.0`  |
| `%`      | Modulus (Remainder) | `10 % 3`  | `1`    |
| `**`     | Exponent (Power)    | `2 ** 3`  | `8`    |
| `//`     | Floor Division      | `10 // 3` | `3`    |

```python
x=10
y=5
print("addition:-",x+y)
print("subtraction:-",x-y)
print("multiplication:-",x*y)
print("division:-",x/y)
print("modulus",x%y)
print("exponent",x**y)
print("floor division",x//y)
```

### ⏺️ Assignment Operators
* Used to assign or update values in variables.
#### Assignment Operators Table
| Operator | Name                    | Example   | Same As      |
| -------- | ----------------------- | --------- | ------------ |
| `=`      | Assign                  | `x = 5`   | `x = 5`      |
| `+=`     | Add and Assign          | `x += 3`  | `x = x + 3`  |
| `-=`     | Subtract and Assign     | `x -= 3`  | `x = x - 3`  |
| `*=`     | Multiply and Assign     | `x *= 3`  | `x = x * 3`  |
| `/=`     | Divide and Assign       | `x /= 3`  | `x = x / 3`  |
| `%=`     | Modulus and Assign      | `x %= 3`  | `x = x % 3`  |
| `**=`    | Exponent and Assign     | `x **= 3` | `x = x ** 3` |
| `//=`    | Floor Divide and Assign | `x //= 3` | `x = x // 3` |

```python
#assignment operator
x=10
print("Assign=",x)

#add and assign
x += 5
print("after +=",x)

#subtract and assign
x -= 3
print("after -=",x)

#multiply and assign
x *= 2
print("after *=",x)

# divide and assign
x /= 4
print("after /=",x)

# modulus and assign
x %= 3
print("after %=",x)

# exponent and assign
x **= 2
print("after **=",x)

#floor and assign
x//=2
print("after //=",x)
```

### ⏺️ Comparision Operator
* Used to compare two values and give result as True or False.
#### Comparision Operator Table
| Operator | Name                  | Example  | Output  |
| -------- | --------------------- | -------- | ------- |
| `==`     | Equal to              | `5 == 5` | `True`  |
| `!=`     | Not Equal to          | `5 != 3` | `True`  |
| `>`      | Greater Than          | `5 > 3`  | `True`  |
| `<`      | Less Than             | `5 < 3`  | `False` |
| `>=`     | Greater Than Equal To | `5 >= 5` | `True`  |
| `<=`     | Less Than Equal To    | `3 <= 5` | `True`  |

```python
a = 10
b = 5

print(a == b)  
print(a != b)  
print(a > b)   
print(a < b)   
print(a >= b)  
print(a <= b)  
```

### ⏺️ Logical Operator
* Used to combine conditions.
#### Logical Operators Table 
| Operator | Name        | Example          | Output  |
| -------- | ----------- | ---------------- | ------- |
| `and`    | Logical AND | `True and False` | `False` |
| `or`     | Logical OR  | `True or False`  | `True`  |
| `not`    | Logical NOT | `not True`       | `False` |

```python
a =True
b =False
#logical And
print("AND operator",a and b)

#logical OR
print("OR operator",a or b)

#logical NOT
print("not operator", not a)
```

### ⏺️ Identity Operator 
* Used to check whether two variables are the same object or not.
#### Identity Operator Table
| Operator | Name                                                   | Example      | Output       |
| -------- | ------------------------------------------------------ | ------------ | ------------ |
| `is`     | Returns True if both variables are the same object     | `a is b`     | `True/False` |
| `is not` | Returns True if both variables are not the same object | `a is not b` | `True/False` |

```python
a=[1,2,3]
b=a
c=[1,2,3]
#is operator
print("a is b:",a is b)

#is not operator
print("a is not c :",a is not c)
```

### ⏺️ Membership Operators
* Used to check whether a value is present in a sequence or not.
#### Membership Operators Table
| Operator | Name                                             | Example              | Output |
| -------- | ------------------------------------------------ | -------------------- | ------ |
| `in`     | Returns True if value is present in sequence     | `'a' in "apple"`     | `True` |
| `not in` | Returns True if value is not present in sequence | `'z' not in "apple"` | `True` |

```python
text = "python"
#in operator
print("'p' in text :",'p' in text)

# not in 
print("'z' not in text : ",'z' not in text)
```

### ⏺️ Bitwise Operators
* Used to perform operations on binary numbers (bits).
#### Bitwise Operators Table
| Operator | Name        | Example    | Output   |     
| -------- | ----------- | ---------- | -------- | 
| `&`      | Bitwise AND | `5 & 3`    | `1`      | 
|  `\|`    | Bitwise OR  | `5 \| 3`   | `7`      |
| `^`      | Bitwise XOR | `5 ^ 3`    | `6`      |     
| `~`      | Bitwise NOT | `~5`       | `-6`     |     
| `<<`     | Left Shift  | `5 << 1`   | `10`     |     
| `>>`     | Right Shift | `5 >> 1`   | `2`      |    

```python
a = 5
b = 3

# Bitwise AND
print("AND (&) :", a & b)

# Bitwise OR
print("OR (|) :", a | b)

# Bitwise XOR
print("XOR (^) :", a ^ b)

# Bitwise NOT
print("NOT (~) :", ~a)

# Left Shift
print("Left Shift (<<) :", a << 1)

# Right Shift
print("Right Shift (>>) :", a >> 1)
```
